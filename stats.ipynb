```python
import pandas as pd

filename = "results.csv"
try:
    df = pd.read_csv(filename, encoding='utf-8-sig')
    
    print(f"{len(df)} candidați încărcați cu succes.\n")
    
    print("Preview:")
    print(df.head())

    df['ro_written'] = pd.to_numeric(df['ro_written'], errors='coerce')
    df['ro_remark']  = pd.to_numeric(df['ro_remark'],  errors='coerce')
    df['ro_final']   = pd.to_numeric(df['ro_final'],   errors='coerce')
    # df['average']    = pd.to_numeric(df['average'],    errors='coerce')
    
    # top_students = df[df['average'] >= 9.00]
    # print(f"\nFound {len(top_students)} students with an average of 9.00 or higher.")
    # print(df[5:10])

except FileNotFoundError:
    print(f"Fișierul '{filename}' nu a fost găsit.")
except Exception as e:
    print(f"Eroare: {e}")
```

# Contestații română

```python
list = df[df['ro_written'] >= 9.95]

num_good = 0
num_raised = 0 
num_equal = 0
num_lowered = 0
lower = []
for i, c in list.iterrows():
    if pd.isna(c['ro_remark']):
        continue
    if c['ro_remark'] == 10:
        num_good += 1
    elif c['ro_remark'] > c['ro_written']:
        num_raised += 1
    elif c['ro_remark'] == c['ro_written']:
        num_equal += 1
    else:
        num_lowered += 1
        lower.append(c['ro_remark'])

print(num_good, num_raised, num_equal, num_lowered)
print(lower)
```
